# Automation

The word _Automation_ means to make the computer do the stuff that had previously done by hand.

car_factory.jpeg

Here is an example. Let's say you want to do print a statement. You could do this once like this

```python
print("automate the boring stuff")
```

If you wanted to print it multiple times you can save typing you can cut and paste this and do it a few times:

```python
print("automate the boring stuff")
print("automate the boring stuff")
print("automate the boring stuff")
```

Now you want to print it 6 times you can cut the code above and paste it twice.

```python
print("automate the boring stuff")
print("automate the boring stuff")
print("automate the boring stuff")
print("automate the boring stuff")
print("automate the boring stuff")
print("automate the boring stuff")
```

But you are still doing manual work so a better way is to write a loop

```python
for i in range(6):
    print("automate the boring stuff")
```

When you are building things it always makes sense to do things as efficiently as you possibly can. At the beginning of the 20th century there were studies called **Time and Motion Studies** where they time how long it would take to do jobs to make the factories more efficient.
